### Report

To improve the team with keeping the payroll sufficient, a clustering analysis was done looking at NBA player performance and salary data for the 2025-26 season. The goal was to find players who produced strong results in relation to their salary, or how much it would cost to have them. These players were identified as the best fits, avoiding players that were overpaid or too poor of an athlete.

Two datasets were combined looking at the salaries of NBA players as well as their performance statistics. In this analysis, points per game and total rebounds were assessed to look at how much each player contributed to their team. Then, salary was used to see which players are underpaid versus overpaid in relation to their skill/contribution to the team.

The data was standardized so the big salary numbers would not overpower the metrics used for the features like points per game and assists.

The data was clustered to group players into categories based on how similar their performance statistics and salaries were. To find a good amount of clusters for the analysis, the Elbow Method and Silhouette Score were used and it was found that there were 3 clusters that showed the best results.

### Mid, High and Low Performers with Salary
These 3 clusters showed high performing and high salary players, mid-range salaries and mid-range performers, and low performing and lower salary players. 
![Final Clusters](final.png)

To find the target players, the average salary was calculated in each cluster. Then, each player's salary was compared to the average salary in their cluster based on their performance (high, medium, low). The players whose salaries were lower than the average in their cluster were assigned as underpaid. Players whose salaries were higher than their cluster's performance average were assigned as overpaid.

From this, three categories of players were found: 1: Not Recommended These players are considered overpaid when compared to the players in their cluster (similar performance). 2: Strong Target Players These players are considered underpaid when compared to players in their cluster (similar performance). 3: Backup Options These players are close to the average salary of their cluster (where players perform similarly), but they still have good performance.
